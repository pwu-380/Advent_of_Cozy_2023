# Part 1

```python
def hand_type_strength(hand):   
    labels = {}
    
    for card in hand:
        if card in labels:
            labels[card] += 1
        else:
            labels[card] = 1
            
    pair = False
    triple = False
    
    for label in labels:
        if labels[label] == 5:
            return 6
        elif labels[label] == 4:
            return 5
        elif labels[label] == 3:
            if pair:
                return 4
            triple = True
        elif labels[label] == 2:
            if triple:
                return 4
            elif pair:
                return 2
            pair = True
        
    if triple:
        return 3
    elif pair:
        return 1
    else:
        return 0


with open('day7.txt') as f:
    hands = [[] for x in range(7)]
    
    for line in f.readlines():
        hand, bid = line.split(" ")
        
        handscore = 0
        face_value = {"T": 10, "J": 11, "Q": 12, "K": 13, "A": 14}
        
        for i, card in enumerate(hand):
            score = int(card) if card.isdigit() else face_value[card]
            handscore += score * 15**(4-i)
                    
        hands[hand_type_strength(hand)].append((handscore, int(bid)))
                
rank = 0
winnings = 0

for hand_type in hands:
    for hand in sorted(hand_type):
        rank += 1
        winnings += rank * hand[1]
        
print(winnings)
```

# Part 2

```python
def hand_type_strength(hand):   
    labels = {}
    
    for card in hand:
        if card in labels:
            labels[card] += 1
        else:
            labels[card] = 1
            
    pair = False
    triple = False
    
    for label in labels:
        if label != "J":
            if labels[label] == 5:
                return 6
            elif labels[label] == 4:
                return 6 if "J" in labels else 5
            elif labels[label] == 3:
                if pair:
                    return 4
                triple = True
            elif labels[label] == 2:
                if triple:
                    return 4
                elif pair:
                    return 4 if "J" in labels else 2
                pair = True
        
    if triple:
        return 4 + labels["J"] if "J" in labels else 3
    elif pair:
        if "J" in labels:
            if labels["J"] == 1:
                return 3
            else:
                return 3 + labels["J"]
        else:
            return 1
    else:
        if "J" in labels:
            if labels["J"] in (4, 5):
                return 6
            elif labels["J"] in (2, 3):
                return labels["J"] * 2 - 1
            else:
                return 1
        else:
            return 0


with open('day7.txt') as f:
    hands = [[] for x in range(7)]
    
    for line in f.readlines():
        hand, bid = line.split(" ")
        
        handscore = 0
        face_value = {"T": 10, "J": 1, "Q": 12, "K": 13, "A": 14}
        
        for i, card in enumerate(hand):
            score = int(card) if card.isdigit() else face_value[card]
            handscore += score * 15**(4-i)
                    
        hands[hand_type_strength(hand)].append((handscore, int(bid)))
                
rank = 0
winnings = 0

for hand_type in hands:
    for hand in sorted(hand_type):
        rank += 1
        winnings += rank * hand[1]
        
print(winnings)
    
```
